# **Hemograma**

É o exame mais solicitado para o auxílio do diagnóstico ou prognóstico das doenças. Pode ser utilizado como exame pré-operatório e nos exames de check-up (NAOUM; NAOUM, 2008; GROTTO, 2009).

O exame é dividido em eritrograma, leucograma e plaquetograma. No eritrograma são obtidos os índices hematimétricos: VCM (volume corpuscular médio), HCM (hemoglobina corpúscular média, CHCM (concentração da hemoglobina corpúscular média) e RDW (Red Cell Distribution Width) (NAOUM; NAOUM, 2008; GROTTO, 2009; BUDAG; CADORE, 2013).

No leucograma será observado os parâmetros da contagem total de leucócitos (CTL) e a contagem diferencial de leucócitos (CDL). O plaquetograma tem a análise quantitativa das plaquetas e para o seu tamanho será utilizado o VPM (volume plaquetário médio) e o PDW (coeficiente de variação do volume plaquetário) (NAOUM; NAOUM, 2008; GROTTO, 2009; BUDAG; CADORE, 2013).

# **Hemácias e leucócitos**




**Hemácias, eritrócitos ou glóbulos vermelhos**

As hemácias são formadas na medula óssea, sendo seu precursor as células mieloides (eritroblastos). São discoidais e não possuem núcleo. São as células sanguíneos mais aundantes no organismo. Sua principal característica e função é o transporte de gases respiratórios no organismo devido às hemoglobinas (MARTY; MARTY, 2015).

```python
he_paciente = float (input('Valor das hemácias do paciente (milhões/mm3):'))
```

**Leucócitos ou glóbulos brancos**

São células com várias formas globulares, nucleadas e que não possuem hemoglobinas. Fazem parte da imunidade do organismo humano (MARTY; MARTY, 2015).

São classificados em granulócitos (neutrófilos, basófilos e eosinófilos), apresentando várias formas de núcleo e vários grânulos em volta de seu citoplasma e agranulócitos (monócitos e linfócitos), com núcleo arredondado e sem granulações no citoplasma (MARTY; MARTY, 2015).

Os neutrófilos realizam a fagocitose de microrganismos e outras substâncias; os basófilos liberam a histamina e a heparina; os eosinófilos participam do combate a parasitas e dos processos alérgicos, assim como os basófilos; os linfócitos produzem anticorpos e agentes que atuam para destruição dos microrganismos, auxiliam nas alergias, rejeição de enxertos, controlam os tumores e regularizam o sistema imune; os monócitos são células que se convertem em macrófagos nos tecidos, sendo os macrófagos células com alta capacidade de fagocitose (MARTY; MARTY, 2015).

# **Hemoglobina**

É a proteína mais abundante nas hemácias, sendo responsável pela coloração avermelhada destas células. Sua principal função é o transporte dos gases respiratórios O2 e CO2 pelo corpo (MARTY; MARTY, 2015).



```python
hb_paciente = float (input('Valor da hemoglobina do paciente (g/dL): '))
```

# **Hematócrito (Ht)**

O hematócrito (Hct ou Ht) é um parâmetro analisado no hemograma e tem por finalidade medir o volume de concentração de hemácias em relação ao sangue total (MONDAL; LOTFOLLAHZADEH, 2022).

Atualmente, graças às inovações tecnológicas, as técnicas laboratoriais estão sendo automatizadas, portanto, os valores de Hct podem ser gerados, em conjunto com o hemograma completo, por um analisador automático (MONDAL; LOTFOLLAHZADEH, 2022).

Há também outros dois métodos para avaliar esse parâmetro: o macro-hematócrito, que é feito no tubo de hematócrito Wintrobe e utiliza a amostra de sangue venoso coletada com tubo contendo ácido etilenodiaminotetracético (EDTA) e o micro-hematócrito, que utiliza um capilar contendo heparina ou não (caso haja sangue com anticoagulante de outros exames hematológicos). Para o micro-hematócrito não há necessidade de grande volume amostral (MONDAL; LOTFOLLAHZADEH, 2022).

O resultado final é dado em porcentagem, e corresponde à razão entre o comprimento da coluna das hemácias e comprimento da coluna do fluído total (hemácias, leucócitos, plaquetas e plasma) multiplicado por 100 (MONDAL; LOTFOLLAHZADEH, 2022). 

O Ht é utilizada para calcular o VCM e o CHCM . Também é função deste exame auxiliar no diagnóstico das anemias e policitemia  e outras enfermidades (MONDAL; LOTFOLLAHZADEH, 2022).


```python
ht_paciente = float (input('Valor do hematócrito do paciente (%): '))
```

# **Cálculo dos índices hematimétricos**

Os índices hematimétricos são parâmetros que auxiliam na interpretação do hemograma e são úteis na classificação das anemias, pois, avaliam o formato, tamanho e o volume dos eritrócitos. Estes índices são o VCM, HCM, CHCM e o RDW (SOUZA et al., 2013).

---

**VCM (Volume corpuscular médio)** - representa a média do volume dos eritrócitos, podendo ser utilizado na classificação de anemias e que pode evidenciar alterações próprias de doenças eritrocíticas. Seu valor é expresso em fL (fentolitro) (SOUZA et al., 2013; MARTY; MARTY, 2015).


# *VCM = (hematócrito / nº de hemácias) x 10  
---

**HCM (Hemoglobina corpuscular média)** - determina a média do conteúdo de hemoglobina por hemácia e pode ser calculado por métodos automatizados ou manuais. Seu resultado é expresso em pg (picograma) (SOUZA et al., 2013; MARTY; MARTY, 2015). 

# *HCM = (hemoglobina / nº de hemácias) x 10 

---

**CHCM (Concentração de hemoglobina corpuscular média)** - representa a média da concentração de hemoglobina em volume sanguíneo. É expresso em porcentagem (SOUZA et al., 2013; MARTY; MARTY, 2015).


# *CHCM = (hemoglobina / hematócrito) x 100

---

**RDW (Red Cell Distribution Width)**, isto é, amplitude de distribuição de células vermelhas) - indica anisocitose (alteração de tamanho) dos eritrócitos, representando a porcentagem de variação dos volumes obtidos (SOUZA et al., 2013; MARTY; MARTY, 2015).
















```python
vcm_paciente = (ht_paciente / he_paciente) * 10
print(f'O VCM do paciente é {vcm_paciente:.2f} fL')
```

```python
hcm_paciente = (hb_paciente / he_paciente) * 10
print(f'O HCM do paciente é {hcm_paciente:.2f} pg')
```

```python
chcm_paciente = (hb_paciente / ht_paciente) * 100
print(f'O CHCM do paciente é {chcm_paciente:.2f} g/dL')
```

```python
rdw_paciente = float (input('Valor do RDW do paciente (%)? '))
```

# **Análise quantitativa do número de hemácias**

```python
if he_paciente > ref_he_max:
  print ('Paciente apresenta policitemia.')
elif he_paciente < ref_he_min:
  print ('Paciente apresenta anemia.')
else:
  print ('Paciente sem alterações no número de hemácias.')
```

# **Análise dos índices hematimétricos**

```python
if vcm_paciente > ref_vcm_max:
  print ('O paciente apresenta macrocitose.')
elif vcm_paciente < ref_vcm_min:
  print ('O paciente apresenta microcitose.')
else:
  print ('Paciente sem alterações no VCM.')
```

```python
if hcm_paciente > ref_hcm_max:
  print ('O HCM do paciente está aumentado.')
elif hcm_paciente < ref_hcm_min:
  print ('O HCM do paciente está diminuído.')
else:
  print ('Paciente sem alterações no HCM.')
```

```python
if chcm_paciente > ref_chcm_max:
  print ('Paciente apresenta hipercromia.')
elif chcm_paciente < ref_chcm_min:
  print ('Paciente apresenta hipocromia.')
else:
  print ('Paciente sem alterações no CHCM.')
```

```python
if rdw_paciente > ref_rdw_max:
  print ('Paciente apresenta anisocitose.')
else:
  print ('Paciente sem alterações no RDW.')
```
